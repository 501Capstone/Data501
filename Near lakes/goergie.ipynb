```python
import pandas as pd
import numpy as np

df = pd.read_csv("gregoire_river_cleaned.csv")
```

```python
df
```

```python
rows_mg = df[df["unit_symbol"] == "mg/l"]

rows_mg
```

```python
dfunits = df.copy()

dfunits["vanadium_total_recoverable"] = pd.to_numeric(
    dfunits["vanadium_total_recoverable"], errors="coerce"
)

condition = dfunits['unit_symbol'] == 'mg/l'
dfunits.loc[condition, 'vanadium_total_recoverable'] *= 1000
dfunits.loc[condition, 'unit_symbol'] = 'µg/l'

dfunits
```

```python
rows_mg = dfunits[dfunits["unit_symbol"] == "mg/l"]

rows_mg
```

```python
df_clean = dfunits.drop(columns=[
    "parameter",
    "source_file",
    "sample_timestamp",
    "measurement_value_raw",
    "station_name",
    "method",
    "sampling_number",
    "sample_number",
    "sample_location",
    "sample_type"

])
df_clean
```

```python
rows_mg = df_clean[df_clean["unit_symbol"] == "mg/l"]

rows_mg
```

```python
header_indices = df_clean[df_clean['site_name'] == 'Site Name'].index.tolist()
print(header_indices)
print(len(header_indices))
```

```python
starts = [8, 17, 23, 33, 43, 53, 60, 70, 79, 88, 97, 106, 115, 124, 137, 154, 163]

dfs = []

for i in range(len(starts)):
    
    start = starts[i]
    
    if i < len(starts) - 1:
        end = starts[i+1]
        df_split = df_clean.iloc[start:end]
    else:
        df_split = df_clean.iloc[start:]   # last section
    
    dfs.append(df_split)
```

```python
#clean_dfs = []

for d in dfs:
    
    chem_name = d.iloc[0, -1]   # chemical name from header row
    
    d.rename(columns={d.columns[-1]: chem_name}, inplace=True)
    
    d.drop(index=d.index[0], inplace=True)  # remove header row
```

```python
"""for i, d in enumerate(clean_dfs):
    print(f"\nDataFrame {i}")
    print(d)"""
dfs[14].head()
```

```python
# Aluminum
dfs[16] = dfs[16].rename(columns={dfs[16].columns[-1]: "Aluminum Dissolved"})
aluminum_df = dfs[16][["sampling_timestamp", "Aluminum Dissolved"]]

dfs[15] = dfs[15].rename(columns={dfs[15].columns[-1]: "Aluminum Total Recoverable"})
aluminum_total_df = dfs[15][["sampling_timestamp", "Aluminum Total Recoverable"]]

# Calcium
dfs[14] = dfs[14].rename(columns={dfs[14].columns[-1]: "Calcium Dissolved"})
calcium_df = dfs[14][["sampling_timestamp", "Calcium Dissolved"]]

dfs[13] = dfs[13].rename(columns={dfs[13].columns[-1]: "Calcium Total"})
calcium_total_df = dfs[13][["sampling_timestamp", "Calcium Total"]]

# Copper
dfs[12] = dfs[12].rename(columns={dfs[12].columns[-1]: "Copper Dissolved"})
copper_df = dfs[12][["sampling_timestamp", "Copper Dissolved"]]

dfs[11] = dfs[11].rename(columns={dfs[11].columns[-1]: "Copper Total Recoverable"})
copper_total_df = dfs[11][["sampling_timestamp", "Copper Total Recoverable"]]

# Lead
dfs[10] = dfs[10].rename(columns={dfs[10].columns[-1]: "Lead Dissolved"})
lead_df = dfs[10][["sampling_timestamp", "Lead Dissolved"]]

dfs[9] = dfs[9].rename(columns={dfs[9].columns[-1]: "Lead Total Recoverable"})
lead_total_df = dfs[9][["sampling_timestamp", "Lead Total Recoverable"]]

# Nickel
dfs[8] = dfs[8].rename(columns={dfs[8].columns[-1]: "Nickel Dissolved"})
nickel_df = dfs[8][["sampling_timestamp", "Nickel Dissolved"]]

dfs[7] = dfs[7].rename(columns={dfs[7].columns[-1]: "Nickel Total Recoverable"})
nickel_total_df = dfs[7][["sampling_timestamp", "Nickel Total Recoverable"]]

# Nitrogen
dfs[6] = dfs[6].rename(columns={dfs[6].columns[-1]: "Nitrogen Total"})
nitrogen_df = dfs[6][["sampling_timestamp", "Nitrogen Total"]]

# Phosphorus
dfs[3] = dfs[3].rename(columns={dfs[3].columns[-1]: "Phosphorus Total Dissolved"})
phosphorus_df = dfs[3][["sampling_timestamp", "Phosphorus Total Dissolved"]]

dfs[2] = dfs[2].rename(columns={dfs[2].columns[-1]: "Phosphorus Total"})
phosphorus_total_df = dfs[2][["sampling_timestamp", "Phosphorus Total"]]

# Vanadium
dfs[0] = dfs[0].rename(columns={dfs[0].columns[-1]: "Vanadium Dissolved"})
vanadium_df = dfs[0][["sampling_timestamp", "Vanadium Dissolved"]]

# Turbidity
dfs[1] = dfs[1].rename(columns={dfs[1].columns[-1]: "Turbidity"})
turbidity_df = dfs[1][["sampling_timestamp", "Turbidity"]]

# pH
dfs[4] = dfs[4].rename(columns={dfs[4].columns[-1]: "pH"})
ph_df = dfs[4][["sampling_timestamp", "pH"]]

# Oxygen BOD
dfs[5] = dfs[5].rename(columns={dfs[5].columns[-1]: "OXYGEN BIOCHEMICAL DEMAND"})
oxygen_bod_df = dfs[5][["sampling_timestamp", "OXYGEN BIOCHEMICAL DEMAND"]]
```

```python
calcium_df
```

```python
GREGOIRE = pd.concat(
    [
        aluminum_df,
        aluminum_total_df,
        calcium_df,
        calcium_total_df,
        copper_df,
        copper_total_df,
        lead_df,
        lead_total_df,
        nickel_df,
        nickel_total_df,
        nitrogen_df,
        phosphorus_df,
        phosphorus_total_df,
        vanadium_df,
        turbidity_df,
        ph_df,
        #oxygen_sat_df,
        oxygen_bod_df
    ],
    axis=0
)


GREGOIRE ["Site Name"] = "gregoire LAKE"
GREGOIRE ["Unit Symbol"] = "µg/l"

GREGOIRE = GREGOIRE .groupby("sampling_timestamp", as_index=False).first()

GREGOIRE ["sampling_timestamp"] = pd.to_datetime(
    GREGOIRE ["sampling_timestamp"], errors="coerce"
)
```

```python
GREGOIRE  = GREGOIRE .sort_values(by="sampling_timestamp")
GREGOIRE 
```
