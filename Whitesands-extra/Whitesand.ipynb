```python
import pandas as pd

aluminum = pd.read_csv("WHITESANDLAKE_Aluminum.csv", encoding="latin1")
calcium = pd.read_csv("WHITESANDLAKE_Calcium.csv", encoding="latin1")
copper = pd.read_csv("WHITESANDLAKE_Copper.csv", encoding="latin1")
lead = pd.read_csv("WHITESANDLAKE_Lead.csv", encoding="latin1")
nickle = pd.read_csv("WHITESANDLAKE_Nickle.csv", encoding="latin1")
phosphorus = pd.read_csv("WHITESANDLAKE_Phosphorus.csv", encoding="latin1")
vanadium = pd.read_csv("WHITESANDLAKE_Vanadium.csv", encoding="latin1")
```

```python

 #convert timestamp
aluminum["Sampling Timestamp"] = pd.to_datetime(aluminum["Sampling Timestamp"], errors="coerce")

# drop unnecessary columns
aluminum = aluminum.drop(columns=[
    "Sample Timestamp",
    "Station Name",
    "Depth",
    "Sampling Number",
    "Sample Number"
])

# clean units
#aluminum["Unit Symbol"] = aluminum["Unit Symbol"].str.replace("u","µ")"""
aluminum.head(5)
```

```python

calcium["Sampling Timestamp"] = pd.to_datetime(
    calcium["Sampling Timestamp"], errors="coerce"
)

# drop unnecessary columns
calcium = calcium.drop(columns=[
    "Sample Timestamp",
    "Station Name",
    "Depth",
    "Sampling Number",
    "Sample Number"
])

calcium["Calcium Dissolved"] = calcium["Calcium Dissolved"] * 1000

calcium["Unit Symbol"] = calcium["Unit Symbol"].str.replace("m","µ")

calcium.head(5)
```

```python

import numpy as np
aluminum_df = aluminum[["Sampling Timestamp", "Aluminum Dissolved"]]
calcium_df = calcium[["Sampling Timestamp", "Calcium Dissolved"]]
copper_df = copper[["Sampling Timestamp", "Copper Dissolved"]]
lead_df = lead[["Sampling Timestamp", "Lead Dissolved"]]
nickle_df = nickle[["Sampling Timestamp", "Nickel Dissolved"]]
phosphorus_df = phosphorus[["Sampling Timestamp", "Phosphorus Total Dissolved"]]
vanadium_df = vanadium[["Sampling Timestamp", "Vanadium Dissolved"]]

df = pd.concat(
    [
        aluminum_df,
        calcium_df,
        copper_df,
        lead_df,
        nickle_df,
        phosphorus_df,
        vanadium_df
    ],
    axis=0
)
```
