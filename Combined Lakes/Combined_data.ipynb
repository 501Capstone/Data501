```python
import pandas as pd
FarLakes = pd.read_csv("FarLakes_Data.csv")
MidLakes = pd.read_csv("MidLake_Data.csv")
NearLakes = pd.read_csv("NearLake_Data.csv")
```

```python
NearLakes.info()
```

```python
df = pd.concat([FarLakes, MidLakes, NearLakes])
```

```python
df["Sampling Timestamp"] = df["Sampling Timestamp"].fillna(df["sampling_timestamp"])
df = df.drop(columns=["sampling_timestamp"])
```

```python
df["Sampling Timestamp"] = pd.to_datetime(
    df["Sampling Timestamp"], errors="coerce"
)
df = df.sort_values(by="Sampling Timestamp")
```

```python
df = df.groupby(
    ["Site Name","Sampling Timestamp"],
    as_index=False
).max()
```

```python
df
```

```python
#df["Sampling Date"] = df["Sampling Timestamp"].dt.date
lastcol = df.columns[-1]
move = df.pop(lastcol)
df.insert(1,lastcol,move)
df
```

```python
df.drop(columns = ["Sampling Timestamp"], inplace = True)
df
```

```python
df_fixdate = df.groupby(
    ["Site Name","Sampling Date"],
    as_index=False
).max()
```
