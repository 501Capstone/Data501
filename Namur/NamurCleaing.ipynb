```python
aluminum_df = aluminum[["Sampling Timestamp", "Aluminum Dissolved"]]
aluminum_total_df = aluminum_total[["Sampling Timestamp", "Aluminum Total Recoverable"]]

calcium_df = calcium[["Sampling Timestamp", "Calcium Dissolved"]]
calcium_total_df = calcium_total[["Sampling Timestamp", "Calcium Total"]]

copper_df = copper[["Sampling Timestamp", "Copper Dissolved"]]
copper_total_df = copper_total[["Sampling Timestamp", "Copper Total Recoverable"]]

lead_df = lead[["Sampling Timestamp", "Lead Dissolved"]]
lead_total_df = lead_total[["Sampling Timestamp", "Lead Total Recoverable"]]

nickle_df = nickle[["Sampling Timestamp", "Nickel Dissolved"]]
nickle_total_df = nickle_total[["Sampling Timestamp", "Nickel Total Recoverable"]]

nitrogen_df = nitrogen[["Sampling Timestamp", "Nitrogen Dissolved"]]

phosphorus_df = phosphorus[["Sampling Timestamp", "Phosphorus Total Dissolved"]]
phosphorus_total_df = phosphorus_total[["Sampling Timestamp", "Phosphorus Total"]]

vanadium_df = vanadium[["Sampling Timestamp", "Vanadium Dissolved"]]

turbidity_df = turbidity[["Sampling Timestamp", "Turbidity"]]

ph_df = ph[["Sampling Timestamp", "pH"]]

oxygen_sat_df = oxygen_sat[["Sampling Timestamp", "Oxygen dissolved % saturation"]]

oxygen_bod_df = oxygen_bod[["Sampling Timestamp", "OXYGEN BIOCHEMICAL DEMAND"]]
```

```python


df = pd.concat(
    [
        aluminum_df,
        aluminum_total_df,
        calcium_df,
        calcium_total_df,
        copper_df,
        copper_total_df,
        lead_df,
        lead_total_df,
        nickle_df,
        nickle_total_df,
        nitrogen_df,
        phosphorus_df,
        phosphorus_total_df,
        vanadium_df,
        turbidity_df,
        ph_df,
        oxygen_sat_df,
        oxygen_bod_df
    ],
    axis=0
)


df["Site Name"] = "NAMUR LAKE"
df["Unit Symbol"] = "µg/l"

df= df.groupby("Sampling Timestamp", as_index=False).first()

df["Sampling Timestamp"] = pd.to_datetime(
    df["Sampling Timestamp"], errors="coerce"
)
```

```python
df = df.sort_values(by="Sampling Timestamp")
df.head(10)
```
